```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
from citipy import citipy
from scipy import stats

# Import API key
from api_keys import weather_api_key
weather_api_key
```

## Generate Cities List

```python
#create data frame of random latitudes and longitudes
cities_df=pd.DataFrame({"long":(np.random.uniform(-180,180,1500)), "lat":(np.random.uniform(-90,90,1500))})

#use citipy to find cities near each coordinate

city_names=[]
country_codes=[]
for i in range(0,len(cities_df['lat'])):
    lat = cities_df.iloc[i]['lat']
    long = cities_df.iloc[i]["long"]
    city = citipy.nearest_city(lat,long)
    city_name = city.city_name
    city_names.append(city_name)
    country_code = city.country_code
    country_codes.append(country_code)
    
cities_df['City'] = city_names
cities_df['Country'] = country_codes
cities_df.head(10)
```

```python
#drop duplicates
cities_clean = cities_df.drop_duplicates('City')
print(len(cities_clean))


cities_clean.head()
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#loop through all the city names , create urls and find the weather
url = "http://api.openweathermap.org/data/2.5/weather?"
units = 'imperial'

dates = []
cloudinesses = []
humidities = []
maxtemps = []
windspeeds = []


for index, row in cities_clean.iterrows():
    city = row['City']
    query_url = url + "appid=" + weather_api_key + "&q=" + city +"&units=" + units
    print("-"*30)
    print(f"Processing City: {city}")
    print(query_url)
    city_weather = requests.get(query_url).json()
    # Add weather info to lists that will become columns, dropping rows if they dont have weather data
    try:
        dates.append(city_weather['dt'])
        cloudinesses.append(city_weather['clouds']['all'])
        humidities.append(city_weather['main']['humidity'])
        maxtemps.append(city_weather['main']['temp_max'])
        windspeeds.append(city_weather['wind']['speed'])
        print("Data Retrieval Successful")
    except KeyError:
        print(f"No data for {city}")
        cities_clean.drop(index, inplace=True)
#adding lists as columns
cities_clean['Date'] = dates
cities_clean['Cloudiness'] = cloudinesses
cities_clean['Humidity'] = humidities
cities_clean['Max Temp'] = maxtemps
cities_clean['Wind Speed'] = windspeeds

print(len(cities_clean))
cities_clean.head(15)


```

## Linear Regression

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#create separate dataframes for north and south
north_df = cities_clean[cities_clean['lat']>0]
south_df = cities_clean[cities_clean['lat']<0]

# run regression for 'nt', north temperature
nt_slope, nt_int, nt_r, nt_p, nt_std_err = stats.linregress(north_df['lat'], north_df['Max Temp'])

nt_fit = nt_slope * north_df['lat'] + nt_int

plt.scatter(north_df['lat'], north_df['Max Temp'])
plt.plot(north_df['lat'], nt_fit, linestyle="--", color='red',label='y={:.2f}x+{:.2f}'.format(nt_slope, nt_int))

plt.title("Temperature vs. Latitude Regression for Northern Hemisphere")
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.legend(fontsize = 14)
plt.show()
plt.savefig("temp_Latitude_north.png")
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# run regression for 'st', south temperature
st_slope, st_int, st_r, st_p, st_std_err = stats.linregress(south_df['lat'], south_df['Max Temp'])

st_fit = st_slope * south_df['lat'] + st_int

plt.scatter(south_df['lat'], south_df['Max Temp'])
plt.plot(south_df['lat'], st_fit, linestyle="--", color='red', label='y={:.2f}x+{:.2f}'.format(st_slope, st_int))

plt.title("Temperature vs. Latitude Regression for Southern Hemisphere")
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.legend(fontsize = 14)
plt.show()
plt.savefig("temp_Latitude_south.png")
```

```python
#As one would expect, temperatures increase as you approach the equator from either hemisphere.
#the regression line increases from lower latitudes and decreases from higher latitudes.
```
